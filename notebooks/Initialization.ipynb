```python
%matplotlib inline

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = tf.random.normal(shape=[100, 100], mean=0.0, stddev=1.0)
X_norm_fan = X * tf.sqrt(2 / 100)
plt.figure(figsize=(8, 8))
plt.imshow(X_norm_fan, interpolation=None)
plt.colorbar()
```

```python
X_norm_var = X * tf.sqrt(2 / tf.reduce_sum(tf.square(X), axis=-1, keepdims=True))
plt.figure(figsize=(8, 8))
plt.imshow(X_norm_var, interpolation=None)
plt.colorbar()
```

```python
col_var = tf.sqrt(2 / tf.reduce_sum(tf.square(X_norm_var), axis=-1, keepdims=True))
tf.reduce_mean(col_var).numpy()
```

```python
col_var = tf.sqrt(2 / tf.reduce_sum(tf.square(X_norm_fan), axis=-1, keepdims=True))
tf.reduce_mean(col_var).numpy()
```

```python
X_norm_diff = X_norm_fan - X_norm_var
plt.figure(figsize=(8, 8))
plt.imshow(X_norm_diff, interpolation=None, cmap='plasma')
plt.colorbar()
```
