# CoordConv

As additional features to the environment we include each grid cell's x, y and r coordinates to allow convolutions to learn better coordinate transformations.

## References

- http://arxiv.org/abs/1807.03247

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import gym
import sys
import numpy as np
import gym_tool_use
import matplotlib.pyplot as plt
import seaborn as sns
import pyoneer.rl as pyrl
```

```python
env = gym.make("TrapTube-v0")
env = pyrl.wrappers.ObservationCoordinates(env)
# env = pyrl.wrappers.ObservationNormalization(env)
observation = env.reset()
observation.shape, observation.dtype
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 6))
for i, ax in enumerate(axes.ravel()):
    im = ax.imshow(observation[..., i])
    fig.colorbar(im, ax=ax)
plt.tight_layout()
```
